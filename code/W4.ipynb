```r
# Зобразимо розподіл та площу під кривою:
library(ggplot2)
ggplot(data.frame(x = c(-3,3)), aes(x)) +
    geom_vline(xintercept = -1.74, linetype=2, colour="blue") +
    stat_function(fun = dnorm, colour="blue") +
    geom_area(stat = "function", fun = dnorm, fill = "red", xlim = c(-1.74,3))
```

```r
# Знайдемо ймовірність отримати z-значення -1.74 або більше:
pnorm(-1.74, lower.tail = FALSE)
```

Розглянемо інший приклад. Припустимо, що частка всіх
студентів університету, які вживали енергетики протягом
останніх 6 місяців становить р = 0.40. Для вибірки n = 200
студентів, яка ймовірність, що відсоток тих, хто вживав
енергетики протягом останніх 6 місяців менша ніж 32%?
n = 200 p = 0.40
P(p < 0.32) - ?

Перевіримо умови, які потрібні для застосування центральної
граничної теореми:
np = 200 × 0.40 = 80 ≥ 15
n(1-p) = 200 × (1 - 0.40) = 120 ≥ 15
Умови виконуються.
Обчислимо
середньоквадратичне
відхилення
вибіркового
розподілу:

s(p) = 0.0346 
z-значення: −2.31

```r
# Зобразимо розподіл та площу під кривою:
ggplot(data.frame(x = c(-3,3)), aes(x)) +
geom_vline(xintercept = -2.31, linetype=2, colour="blue")+
stat_function(fun = dnorm, colour="blue") +
geom_area(stat = "function", fun = dnorm, fill = "red", xlim = c(-3,-2.31))
```

```r
# Знайдемо ймовірність отримати z-значення -2.31 або менше:
pnorm(-2.31)
```

Ймовірність, що у вибірці з 200 випадковим чином обраних

останніх 6 місяців буде менша ніж 32% становить 1%.
